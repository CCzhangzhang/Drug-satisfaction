```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('filtered_original.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y')

# Extract year and month into separate columns
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# Plotting the distribution of years
data['Year'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Data Entries Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Entries')
plt.show()

# Plotting the distribution of months
data['Month'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Data Entries Over Months')
plt.xlabel('Month')
plt.ylabel('Number of Entries')
plt.show()

# Optional: Age distribution if needed
if 'Age' in data.columns:
    data['Age'].hist(bins=20)
    plt.title('Age Distribution')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
```
